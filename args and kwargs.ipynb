```python
# What do you mean by *args and **kwargs?
```

```python
def func(*args):
    for i in args:
        print(i)
    print(args)
```

```python
func(1,2,3,4)
```

```python
def fun(**kwargs):
    for i in kwargs:
        print(i, kwargs[i])
    print(kwargs)
```

```python
fun(a=1, b=2, c=3)
```
